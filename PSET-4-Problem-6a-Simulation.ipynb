```python
import numpy as np
from math import sin, cos
```

```python
sin_out, cos_out, T_out = [], [], []
for i in range(0,1000000):
  T = np.random.sample()*5000 - 5000//2
  T_out.append(T)
  sin_out.append(sin(T))
  cos_out.append(cos(T))
```

```python
np.mean(sin_out), np.mean(cos_out), np.mean(T_out)
```
