```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
N = 1000000 # number of simulations to run
p = 0.5 # probability of getting heads
R_list = []
R2_list = []
```

# (a) Expected Value Calculations

```python
for i in range(0,N):
  flips = [] # flip results
  R = 0 # sum of indicator variables
  for k in range(0,n):
    flip = np.random.sample()
    if flip <= p:
      flips.append('H')
    else:
      flips.append('T')
    if k >= 1:
      if flips[k-1] != flips[k]:
        R_i = 1
        R = R + R_i
  R_list.append(R)
  R2_list.append(R**2)
```

```python
np.mean(R_list)
```

```python
n*2*p*(1-p)
```

# (b) Variance

```python
np.mean((R_list - np.mean(R_list))**2)
```

```python
2*n*p*(1-p) + (n-1)*p*(1-p) + (1 - 2*n)*4*p**2*(1-p)**2
```

```python
np.mean(R2_list) # experimental value for E[R^2]
```

```python
2*n*p*(1-p) + (n-1)*p*(1-p) + (n**2 + 1 - 2*n)*4*p**2*(1-p)**2 # theoretical value for E[R^2]
```

```python
np.mean((R_list))**2 # experimental value for E[R]^2
```

```python
4*n**2*p**2*(1-p)**2 # theoretical value for E[R]^2
```
