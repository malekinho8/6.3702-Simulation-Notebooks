```python
import numpy as np
```

```python
p = 0.1
R = 20
```

```python
temp = []
for i in range(0,1000000):
  m = 0 # number of roommates
  for r in range(0,R):
    X = np.random.rand()
    if X <= p:
      m += 1
    else:
      continue
  temp.append(m)
```

```python
out = np.array(temp)
```

```python
new = out[(1*(out == 2) + 1*(out == 1)).astype(bool)]
```

```python
len(new)
```

```python
expected_simulation = np.mean(new)
actual_value = (2*(1-p) + 2*(R-1)*p)/(2*(1-p) + (R-1)*p)
```

```python
print(expected_simulation), print(actual_value)
```
