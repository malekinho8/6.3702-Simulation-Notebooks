```python
i = 4
```

```python
n = 10
P_den = 0
p = 0.4
for x in range(0,n+1):
  y = n - x
  add = (1-p)**(n-x)*p**x * (1-p)**(n-y)*p**y
  P_den += add
  if x == i:
    p_c = add
  print(x,y,add)
```
