```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.arange(1,25)
n = np.arange(1,25)
out = np.zeros((len(m),len(n)))
N = 10000 # number of times to run simulation
for i,m_i in enumerate(m):
  for j,n_i in enumerate(n):
    num_diff = 0
    for k in range(N):
      urn = ['r' for i in range(m_i)] + ['w' for j in range(n_i)]
      balls = np.random.choice(urn,(2,),replace=False)
      if balls[0] != balls[1]:
        num_diff += 1
    out[i,j] = num_diff
```

```python
plt.imshow(out/N,origin='lower')
plt.colorbar()
plt.clim([0,1])
plt.show()
```

```python
# theoretical calculation
z = np.zeros((len(m),len(n)))
for i,m_i in enumerate(m):
  for j,n_i in enumerate(n):
    z[i,j] = 2*m_i*n_i/((n_i+m_i)*(n_i+m_i-1))
```

```python
plt.imshow(z,origin='lower')
plt.colorbar()
plt.clim([0,1])
plt.show()
```
