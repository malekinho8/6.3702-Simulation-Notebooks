```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000000
N = 100
Zn = np.random.choice(np.array([3,1/3]),size=(N,n),p=np.array([1/3,2/3]))
```

```python
np.mean()
```

```python
Xn = np.cumprod(Zn,axis=1)
```

```python
np.mean(Xn)
```

```python
np.var(Xn)
```

```python
np.max(Xn)
```

```python
plt.imshow(Xn)
```

```python
plt.plot(Xn[0,:])
plt.xlim([0,20])
```
