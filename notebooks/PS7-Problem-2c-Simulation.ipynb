```python
N = 1000000
D = []
L2_Bover9 = []
L2_Bless9 = []
L_over9 = []
X_over9 = []
B2_over9 = []
B_over9 = []
test = []
for n in range(N):
  A = 9
  B = np.random.uniform(8,10)
  X = B - 8
  if B < 9:
    L = 3
    L2_Bless9.append(L**2)
  else:
    L = np.random.uniform(0,3-X)
    X_over9.append(X)
    L_over9.append(L)
    L2_Bover9.append(L**2)
    B2_over9.append(B**2)
    B_over9.append(B)
    test.append(((3-X)/2)**2)
  D.append(L)
```
