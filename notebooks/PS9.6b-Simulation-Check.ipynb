```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
s = 5
N = 1000000
X = np.random.uniform(-sqrt(3),sqrt(3),(N,))
W = np.random.uniform(-sqrt(3),sqrt(3),(N,))
```

```python
plt.hist(s*W,bins=100)
plt.show()
```

```python
Y = X + s*W
```

```python
plt.hist(Y,bins=100,density=True)
plt.show()
```
