```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
s = 1
N = 100000
n = 10
```

```python
Xa = np.random.normal(0,1,(N,1))
Wa = np.random.normal(0,1,(N,n))
Ya = Xa + s*Wa
```

```python
Ya.shape
```

```python
def h_G(Y,n,s):
  out = []
  for i in range(Y.shape[0]):
    out.append(np.sum(Y[i,:])/(n+s**2))
  return np.array(out)
```

```python
def h_M(Y,n):
  b = np.sort(Ya,axis=1)
  if not n % 2: # if n is even
    k_1 = n // 2
    k_2 = k_1 + 1
    out = 0.5*(b[:,k_1] + b[:,k_2])
  else:
    k = (n - 1)//2
    out = b[:,k]
  return out
```

```python
hg_a = h_G(Ya,n,s)
hm_a = h_M(Ya,n)
```

```python
hga_MSE = np.mean((hg_a - Xa[:,0])**2)
hma_MSE = np.mean((hm_a - Xa[:,0])**2)
```

```python
hga_MSE, hma_MSE
```
