```python
import numpy as np
import matplotlib.pyplot as plt
from math import tan, atan, pi, sqrt, cos, sin
```

```python
X = np.random.uniform(0,1)
Y = np.random.uniform(0,1)
N = 1000000
```

```python
R_vals = []
Th_vals = []
X_vals = []
Y_vals = []
for n in range(N): # perform experiment N times
  X = np.random.uniform(0,1)
  Y = np.random.uniform(0,1)
  R = np.sqrt(X**2 + Y**2)
  Th = atan(Y/X)
  R_vals.append(R)
  Th_vals.append(Th)
  X_vals.append(X)
  Y_vals.append(Y)
```

```python
min(R_vals), min(Th_vals), max(R_vals), max(Th_vals)
```

```python
plt.hist2d(X_vals,Y_vals,bins=100)
plt.colorbar()
```

```python
matrix = plt.hist2d(R_vals,Th_vals,bins=50)
plt.colorbar()
```

```python
th = np.linspace(0,pi/2,100)
r = np.linspace(0,sqrt(2),100)
out = np.zeros((100,100))
for i, ri in enumerate(r):
  for j, t in enumerate(th):
    if ri <= 1:
      out[i,j] = pi/2*ri
    else:
      out[i,j] = pi/2*ri*1/2*cos(t)**2 if t <= pi/4 else pi/2*ri*1/2*sin(t)**2
```

```python
plt.imshow(out.T,origin='lower')
```

```python
r = np.linspace(0,sqrt(2),1000)
t = np.linspace(0,pi/2,1000)
frt = np.zeros((1000,1000))
for i, ri in enumerate(r):
  for j, ti in enumerate(t):
    if ri*cos(ti) <= 1 and ri*cos(ti) > 0 and ri*sin(ti) <= 1 and ri*sin(ti) > 0:
      frt[i,j] = ri
```

```python
plt.imshow(frt.T,origin='lower')
```
