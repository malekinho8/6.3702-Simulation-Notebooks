```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
N = 10000 # number of times to perform experiment
k = np.arange(2,21)
```

```python
outx1 = []
outx2 = []
for ki in k:
  dice = np.arange(1,ki+1)
  test = np.random.choice(dice,(N,n))
  x1 = np.sum(test==1,axis=1)
  x2 = np.sum(test==2,axis=1)
  outx1.append(x1)
  outx2.append(x2)
outx1 = np.array(outx1)
outx2 = np.array(outx2)
```

```python
outx1.shape
```

```python
test = np.random.choice([1,2,3,4],(10000,10000))
x1 = np.sum(test==1,axis=1)
x2 = np.sum(test==2,axis=1)
```

```python
x1, x2
```

```python
fig1, axes = plt.subplots(nrows=4,ncols=4,figsize=(10,8))
idx = 0
for i in range(4):
  for j in range(4):
    axes[i,j].scatter(outx1[idx,:],outx2[idx,:])
    axes[i,j].set_title(f'k = {k[idx]}')
    idx+=1
plt.tight_layout()
plt.show()
```
