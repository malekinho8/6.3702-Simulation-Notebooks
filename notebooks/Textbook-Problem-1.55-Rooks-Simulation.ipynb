```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
board = np.arange(1,65).reshape(8,8)
N = 1000000 # number of times to perform experiment
rook_count = 0
```

```python
for i in range(N):
  rooks = np.random.choice(board.reshape(-1),(8,),replace=False)
  temp = 0
  for rook in rooks:
    if rook in np.diag(board).tolist():
      temp += 1
  if temp == 8:
    rook_count += 1
```

```python
rook_count / N
```
