```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
K = 20
out = []
N2 = []
for n in range(N):
  p = np.random.uniform(0,1)
  B = 0
  for k in range(K):
    test = np.random.uniform(0,1)
    if test <= p:
      B += 1
  out.append(B)
  N2.append(B**2)
```
