```python
result = []
universe = []
for i in range(0,1000000):
  pA = np.random.rand()
  pB = np.random.rand()
  pC = np.random.rand()
  if pA <= p1:
    a = 1
  elif pA > p1 and pA <= (p1 + p2):
    a = 2
  else:
    a = 3
  if pB <= p1:
    b = 1
  elif pB > p1 and pB <= (p1 + p2):
    b = 2
  else:
    b = 3
  if pC <= p1:
    c = 1
  elif pC > p1 and pC <= (p1 + p2):
    c = 2
  else:
    c = 3
  result.append((a,b,c))
  if (b == 3 or (b==2 and (a==3 or c==3))):
    universe.append((a,b,c))
```
