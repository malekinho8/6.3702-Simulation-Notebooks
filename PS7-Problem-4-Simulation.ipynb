```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
p = 1/2
coin = ['H','T']
outK = []
outX = []
X2 = []
X2_Kis1 = []
X_Kis1 = []
XX_Kis1 = []
for n in range(N):
  K = 0
  X = 0
  XX = []
  while True:
    flip = np.random.choice(coin,p=[p,1-p])
    Xk = np.random.uniform(0,3)
    X += Xk
    XX.append(Xk)
    if flip == 'H':
      break
    K += 1
  outK.append(K)
  outX.append(X)
  X2.append(X**2)
  if K == 1:
    XX_Kis1.append(XX)
    X2_Kis1.append(X**2)
    X_Kis1.append(X)
```
